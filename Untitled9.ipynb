```python
import pandas as pd
```

```python
jj=pd.DataFrame({'StepsPerDay':[1200,11000,10000,9000,8000,7000,6000,5000,4000,3000],
                'FitnessScore':[95,91,88,82,78,70,65,60,50,45]})
```

```python
x=jj[['StepsPerDay']]
y=jj['FitnessScore']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
jj=LinearRegression()
```

```python
jj.fit(x_train, y_train)
```

```python
y_pred=jj.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse=mean_squared_error(y_test, y_pred)
```

```python
r2=r2_score(y_test, y_pred)
```

```python
print('mean_squared_error(mse)',mse)
print('r2 score',r2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_test, y_test, color='red', label='actual values')
plt.scatter(x_test, y_pred, color='black',label='predicted values')
plt.xlabel('StepsPerDay')
plt.ylabel('FitnessScore')
plt.legend('')
```
